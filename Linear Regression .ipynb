-Predicting Pizza Price


```python
import pandas as pd
```

```python
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
data = {
    'pizza': [8,10,12],
    'price': [10, 13, 16]
}
```

```python
df = pd.DataFrame(data)
```

```python
%matplotlib inline
plt.xlabel('pizza')
plt.ylabel('price')
plt.scatter(df.pizza,df.price,color='red',marker='+')
```

```python
# Separate features and target
X = df[['pizza']]
Y = df['price']
```

```python
# Fit the linear regression model
new_df = df.drop('price',axis='columns')
new_df
```

```python
price = df.price
price
```

```python
# Create linear regression object
reg = LinearRegression()
reg.fit(new_df,price)
```

```python
# Make predictions
predictions = reg.predict([[30]])
```

```python
print(predictions)
```
