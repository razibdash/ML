```python
import pandas as pd
from sklearn import tree
```

```python
df=pd.read_csv('dataset.csv')
df
```

```python
x=df.iloc[:,:-1]
x
```

```python
y=df.iloc[:,4]
y
```

```python
classify_=tree.DecisionTreeClassifier()
classify_
```

```python
classify_=classify_.fit(x,y)
```

```python
predicted_=classify_.predict([["male",0,'cheap','low']])
predicted_
```
